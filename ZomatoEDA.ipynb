```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('zomato.csv',encoding='latin-1')
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country
```

```python
df_final = pd.merge(df,df_country,on='Country Code',how = "left")
```

```python
df_final.head()
```

```python
df_final.Country.value_counts()
```

```python
#Top 3 Countries that uses Zomato
sns.barplot(x=df_final.Country.value_counts().index[:3],y=df_final.Country.value_counts()[:3],color="red")
```

Observation: Zomato maximum transaction is from India then United States and then United Kingdom


```python
ratings = df_final.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings.head()
```

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,hue = "Rating color",palette = ["blue","red","orange","yellow","green","green"])
```
